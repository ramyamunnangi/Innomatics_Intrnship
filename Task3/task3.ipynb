```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('R:/aspiring_minds_employability_outcomes_2015.xlsx')
df.head()
```

# **1.Violinplot**

 ->It is a Bivariate plot.

 ->It plot both numarical and categorical variables.
 
 ->A violin plot plays a similar role as a box and whisker plot.

**Observation: From the data tier1 college students getting maximum salary and  in those males were more.**

```python
sns.violinplot(x='CollegeTier', y='Salary', data=df,
               hue='Gender', split=True)
```

# **2.relplot**

  ->It is Bivariate plot.
  
  ->It deals with both numarical and categorical varibales.

  ->It is designed to visualize many different statistical relationships. 

**Observation: The students who are from B.Tech/B.E and MCA were getting maximum salaries.**

```python
sns.relplot(
    data=df, 
    y="Salary", x="Degree", hue="Salary" 
    
)
```

# **3.lmplot**

  ->It is a Bivariate plot.
  
  ->It plots only numarical variables.
  
  ->lmplot() combines regplot() and FacetGrid.
  

**Observation: From the data students who scored above 80% in 10th class were getting heighest salaries.**

```python
sns.lmplot(data=df, y="Salary", x="10percentage")
```

# **4.catplot**(kind = bar , swarm, viloin , point)

  ->It is a Bivariate plot.
  
  ->It deals with both categorical and numarical variables.

**Observation: As compared to females males were getting hieghest sarales**

```python
sns.catplot(data=df, x="Gender", y="Salary" , hue="Salary")
```

# **5.Boxenplot**

  -> It is also called as"letter-value-plot"

  -> It is a Bivariate plot.
  
  ->It plots both numarical and categorical variables.
  
  ->Boxen plot provide a better representation of the distribution of the data than boxplot.

**Observation: The students who were getting maximum salary are from 2012 graduation year and the students who were getting least salaries are from 2016 graduation year.**

```python
sns.boxenplot(x="GraduationYear", y="Salary", data=df)
```

# **6.residplot**

  ->It is a Bivariate plot.
  
  ->It contains both numarical and categorical variables.

**Observation : From the data the students who were having above 80% in 10th percentage getting maximum salaries.**

```python
sns.residplot(x="10percentage", y="Salary",data=df )
```

# **7.factorplot**(kind=swarm)

   ->It is a Bivariate plot.

   -> It deals with both numarical and categorical variables.

**Observation: From the data most of the students from B.Tech/B.E , less students from M.Sc and most of the students having above 80% in their collegeGPA.**

```python
g = sns.factorplot(x='Degree', y='collegeGPA',data=df ,kind='swarm') 
                   
```

# **8.PairGrid**

  ->It is a Bivariate plot.
  
  ->It deals with only numarical variables.

**Observation: Collegetier1 students having above 75% collegeGPA.** 

```python
g = sns.PairGrid(df, y_vars="collegeGPA",
                 x_vars=["CollegeTier"] ,  height=5, aspect=.5)
g.map(sns.pointplot, scale=1.3, errwidth=4, color="xkcd:plum")
              
```

**9. Facetgridplot**

 -> It is a Bivariate plot.
 
 -> It deals with only categorical variables.

**Observation :The students from  uttar pradesh were getting more salaries compared to other states .**

```python
g = sns.FacetGrid(df, col="CollegeState", hue="Gender")
g.map(plt.hist, "Salary")
plt.show()
```
